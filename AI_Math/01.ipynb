### 1
$f(x)$的定义域为：$-2\le x\le 1$  
$f(-1) = -2$  
$f(1) = 2$  
$f(x^2) = x^2 + 1$

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-2, 1, 1000)
case1 = [1 if (-2 <= i < 0) else 0 for i in x]
case2 = [1 if (0 <= i <= 1) else 0 for i in x]
y = (x - 1) * case1 + (x + 1) * case2
plt.plot(x, y)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 1000)
case1 = [1 if (i < -1) else 0 for i in x]
case2 = [1 if (-1 <= i <= 1) else 0 for i in x]
case3 = [1 if (1 < i) else 0 for i in x]
y = (-1-2*x) * case1 + (np.sin(x)+1) * case2 + (np.e**(x-1)) * case3
plt.plot(x, y)
plt.show()
```

### 6

定义域：$x \neq 1$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

F = lambda x: (x**2-x+4)/(x-1)

def dfdx(num):
    x = sp.symbols('x')
    dfdx = sp.diff(F(x), x) 
    return dfdx.subs('x', num)

x = np.linspace(-10, 10, 100)
dfdx_y = list(map(dfdx, x))
plt.plot(x, F(x), label=u'f(x)')
plt.plot(x, dfdx_y, label=u'f\'(x)', linestyle='--')

plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)

plt.legend()
plt.show()
```

根据函数图像可知：  
函数在 $(-\infty, -1)$ 和 $(3, +\infty)$ 单调递增  
在 $(-1, 3)$ 单调递减

```python
from scipy import optimize
fun1 = lambda x:(x**2 - x + 4) / (x - 1)
fun2 = lambda x:-((x**2 - x + 4)/(x - 1))

xmin = optimize.minimize(fun1,0)
xmax = optimize.minimize(fun2,0)
print('极小值点和极小值为：',xmin.x[0],xmin.fun)
print('极大值点和极大值为：',xmax.x[0],xmax.fun)
```

### 8

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

F = lambda x: x**4 + 2*x**3 - 3*x**2 + 4*x - 5
df2dx2 = lambda x: 12*x**2 + 12*x - 6

x = np.linspace(-3, 3, 100)
F_y = list(map(lambda i: F(i), x))
dF_y = list(map(lambda i: df2dx2(i), x))

plt.plot(x, F_y, label=u'f(x)')
plt.plot(x, dF_y, label=u'f\'\'(x)', linestyle='--')
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)


plt.legend()
plt.show()

root = optimize.fsolve(df2dx2, 0.3)
root1 = optimize.fsolve(df2dx2, -1.3)
print('根点为：', root1[0], root[0])
```

凹区间为：$[-\infty, -1.37] [0.37, +\infty]$  
凸区间为：$(-1.37, 0.17)$  
拐点为：$-1.37, 0.37$

```python
import sympy as sp

x, y = sp.symbols('x y')
z = sp.E**(y/x)
print('x 的偏导', sp.diff(z, x))
print('y 的偏导',sp.diff(z, y))
print('xy 的二阶偏导',sp.diff(sp.diff(z, x), y))
```
